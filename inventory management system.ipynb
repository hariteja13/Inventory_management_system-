```python
import json
import time
fd=open('Record.json','r')
js=fd.read()
fd.close()
#converting string to dict
record = json.loads(js)
#displaying menu
print("--------------------MENU--------------------")
for key in record.keys():
    print(key,record[key]['Name'],record[key]['Price'],record[key]['Qn'])
print("---------------------------------------------")
print('')
#taking input from user 
ui_name = str(input("Enter your Name   : "))
ui_mail = str(input("Enter mail Id     : "))
ui_ph   = str(input("Enter Phone No    : "))
ui_pr   = str(input("Enter Product ID  : "))
ui_qn   = int(input("Enter Quantitiy       : "))
print("--------------------------------------------")
print('')
#if inventory quantity is => user wants
if (record[ui_pr]['Qn']>= ui_qn):
    print("Name       :",record[ui_pr]["Name"])
    print("Price(Rs)  :",record[ui_pr]["Price"])
    print("Quantity   :",ui_qn)
    print("----------------------------------------------")
    print("Billing    :",ui_qn*record[ui_pr]["Price"],"Rs")
    print("----------------------------------------------")
    #update inventory in dictonary
    record[ui_pr]['Qn'] = record[ui_pr]['Qn'] - ui_qn
    
    #genrating csv transfer details file
    sale = ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["Name"]+","+str(ui_qn)+","+str(record[ui_pr]["Price"])+","+str(ui_qn * record[ui_pr]["Price"])+","+time.ctime()+"\n"
# if inventory is less
else:
    print("Sorry,We are low in stock")
    print("we only have"+str(record[ui_pr]['Qn'])+"quantity.")
    print("-----------------------------")
    
    ch = str(input("Press Y to purchase:"))
    
    #if user wants to purchase the whole inventory
    if(ch =='Y' or ch == 'y'):
        print("--------------------------------------")
        print("Name       :",record[ui_pr]["Name"])
        print("Price(Rs)  :",record[ui_pr]["Price"])
        print("Quantity   :",record[ui_pr]['Qn'])
        print("--------------------------------------")
        print("Billing    :",record[ui_pr]['Qn']* record[ui_pr]["Price"],"Rs")
        print("---------------------------------------")
        
        #update inventory dictonary
        record[ui_pr]['Qn']= 0
        
        #generate csv transaction details 
        sale = ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["Name"]+","+str(record[ui_pr]['Qn'])+","+str(record[ui_pr]["Price"])+","+str(record[ui_pr]['Qn'] * record[ui_pr]["Price"])+","+time.ctime()+"\n"
    #if user pressed any other except Y ot y
    else:
        print("Thankyou for shopping with us !!!")

#conversion invertory dict into string
js = json.dumps(record)

#updating inventory and saving in my records.json
fd = open('Record.json','w')
fd.write(js)
fd.close()

#adding transaction on sales file
fd = open('Sales.txt','a')
fd.write(sale)
fd.close()

print('')
print("--------------------------------")
print("Thankyou , Inventory updated.")
print("--------------------------------")
```
